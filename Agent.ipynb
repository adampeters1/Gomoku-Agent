```python
'''
My Policy/Agent, uses Minimax to minimise possible loss.
'''

# Define the reward values for possible configurations of each line that is passed to it.
def line_rewards(line):
    reward = 0
    
    # positively append reward, based on number of o's.
    reward += line.count("o") * 2
    reward += line.count("oo") * 3
    reward += line.count("ooo") * 4
    reward += line.count("oooo") * 5

    # negatively append reward, based on number of x's.
    reward -= line.count("x") * 3
    reward -= line.count("xx") * 4
    reward -= line.count("xxx") * 6
    reward -= line.count("xxxx") * 8
    
    # Win condition for "o" - meaning my agent wins.
    if("ooooo" in line):
        reward = 1000
        return reward
    
    # Lose condition, for 'x' - your agent wins.
    if("xxxxx" in line):
        reward = -1000
    
    # This reward is unindented so that the value can be retreived when line_rewards is called below:
    return reward 

def append_reward(board, my_turn):
    reward = 0
    
    # This a reuse of your code in cell 1, perform check on all the lines (rows, columns, diagonals, flipped diagonals).
    # The difference being we then use this checked value to append the reward value from line_rewards to "reward"
    for row in board:
        row_str = ''.join(row).lower()
        reward += line_rewards(row_str)
        
    for i in range(0,board.shape[1]):
        col = board[:,i]
        col_str = ''.join(col).lower()
        reward += line_rewards(col_str)
        
    for i in range(0,board.shape[1]):
        diagonal = board.diagonal(offset=i)
        diagonal_str = ''.join(diagonal).lower()
        reward += line_rewards(diagonal_str)
        
    for i in range(0,board.shape[0]):
        diagonal = board.diagonal(offset = i, axis1 = 1, axis2 = 0)
        diagonal_str = ''.join(diagonal).lower()
        reward += line_rewards(diagonal_str)
        
    flipped_board = np.flip(board,axis=1)
    
    for i in (range(flipped_board.shape[0])):
        diagonal = flipped_board.diagonal(offset = i, axis1 = 0, axis2 = 1)
        diagonal_str = ''.join(diagonal).lower()
        reward += line_rewards(diagonal_str)
        
    for i in (range(flipped_board.shape[1])):
        diagonal = flipped_board.diagonal(offset = i, axis1 = 1, axis2 = 0)
        diagonal_str = ''.join(diagonal).lower()
        reward += line_rewards(diagonal_str)

    return reward


def miniMax(board, my_turn = True, last_cost = -1000, depth = 0):
    depth += 1
    
    # As the returns for going over a depth of 2 is diminishing, MD is 1.
    max_depth = 1
    
    # Assign the current reward value
    reward = append_reward(board, not my_turn)
    best_move = None
    best_reward = None
    best_move_row = None
    best_move_col = None

    if(depth > max_depth):
        return best_move, reward
    
    # Changes the reward value, based on the game outcome.
    terminal_status = terminal(board)
    if terminal_status == 'X':
        return best_move, -100
    elif terminal_status == 'O':
        return best_move, 100
    elif terminal_status == 'tie':
        return best_move, 0
    
    # Check free spaces - restatement of your code in cell 3.
    available_moves_idx = np.where(board==' ')
    available_moves = board[available_moves_idx]
    
    # Play a random move at the start, so the middle positions arent always played (most effective, but low-varience.)
    if(len(available_moves_idx[0]) == 224):
        move_idx = random.randint(0, len(available_moves_idx[0]))
        rand_row = available_moves_idx[0][move_idx-1]
        rand_col = available_moves_idx[1][move_idx-1]
        return (rand_row * 15) + rand_col + 1, 0
    
    # Checks all available moves before playing.
    for move_idx in range(len(available_moves_idx[0])):
        move_idx_row = available_moves_idx[0][move_idx]
        move_idx_col = available_moves_idx[1][move_idx]
        # Make a copy of the board.
        local_board = copy.copy(board)
        
        # Set the initial reward value low, play all available spaces, evaluate which would give the best reward. 
        if(my_turn):
            reward = -1000
            local_board[move_idx_row, move_idx_col] = 'O'
            reward = max(reward, miniMax(local_board, False, -1000, depth)[1])
            # In the event there is none, BR = -1000
            if(best_reward == None):
                best_reward = -1000
            # Assign the chosen coordinates, and reward value.
            if(reward >= best_reward):
                best_move_row = move_idx_row
                best_move_col = move_idx_col
                best_reward = reward
            
    return (best_move_row * 15) + best_move_col + 1, best_reward
    
```
